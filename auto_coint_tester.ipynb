## Automatic Cointegration Pair Tester Set Up

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_prices = pd.read_csv('SP500.csv')
#importing csv of all S&P500 companies with their ticker and sector
```

```python
df_prices.head()
```

```python
#Find unique sectors to do cointegration
df_prices['Sector'].unique()
```

```python
# finding all undownloadable companies and removing them later
for i in df_prices['Symbol']:
    yf.download(i,'2020-01-01','2025-12-20')
```

```python
companies_to_remove = ['ATVI','AET','ALXN','AGN','ADS','ABC','APC','ANDV',
                       'ANSS','ANTM','ARNC','BHGE','BLL','BRK.B','BF.B',
                       'COG','CBG','CBS','CELG','CTL','CERN','CHK','XEC',
                       'CTXS','CXO','CSRA','DFS','DISCA','DISCK','DISH','DWDP',
                       'DPS','DRE','ETFC','EVHC','RE','ESRX','FLIR','FL',
                       'FBHS','GPS','GGP','HRS','HCP','HES','JEC','JNPR',
                       'KSU','LLL','LUK','MRO','KORS','MON','MYL','NFX','NLSN',
                       'NBL','JWN','PDCO','PBCT','PKI','PXD','RTN','RHT','SCG',
                       'SNI','SRCL','SYMC','TIF','TWX','TMK','TSS','UTX','VAR',
                       'VIAB','WBA','HCN','WRK','WLTW','WYN','XLNX','XL',]
df_clean = df_prices[~df_prices['Symbol'].isin(companies_to_remove)]
# using all unavailable companies to clean df
```

```python
# making a dictionary to pair companies with their sector
sectors_dict = df_clean.groupby('Sector')['Symbol'].apply(list).to_dict()
```

```python
df_clean.to_csv('SP500_clean.csv')
# export to csv
```
