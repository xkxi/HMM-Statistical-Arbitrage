```python
import pandas as pd
import numpy as np
from hmmlearn.hmm import GaussianHMM
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# read in necessary csv
df_results = pd.read_csv("cointegration_results.csv")
df_prices = pd.read_csv("final_price_clean.csv", index_col=0, parse_dates=True)
```

```python
# incorporate spread calculation from before
BLK_SYF = df_prices[['BLK','SYF']]

BLK_norm = np.log(BLK_SYF['BLK'])
SYF_norm = np.log(BLK_SYF['SYF'])

SYF_norm_c = sm.add_constant(SYF_norm)

model = sm.OLS(BLK_norm,SYF_norm_c).fit()

spread = model.resid
```

```python
# reshape series into numpy array
spread = spread.values.reshape(-1,1)
```

```python
#Defining the model
hmm_model = GaussianHMM(n_components=2, 
                        n_iter=10000) 
                        # init_params="st",
                        # min_covar=1e-6,
                        # random_state=42)
```

```python
# fitting the model
hmm_model.fit(spread)
# Defining hidden states
hidden_states = hmm_model.predict(spread)
```

```python
len(spread[hidden_states ==1])
```

```python
# figure which state is safe and which is risky
for i in range(0,2):
    state_spreads = spread[hidden_states == i]
    print(f"State {i}:")
    print(f"  Mean: {state_spreads.mean():.4f}")
    print(f"  Variance: {state_spreads.var():.6f}")
    print(f"  Count: {len(state_spreads)} days")
```

```python
# calculate z-score
z_score = (spread - spread.mean()) / spread.std()
```

```python
#create trading dataframe for future algorithm
df_trade = pd.DataFrame(index = BLK_SYF.index)
df_trade['Spread'] = spread
df_trade['State'] = hidden_states
df_trade['Z_score'] = z_score

df_trade.head()
```

```python
# iterate through every day to make a decision
#position can be 1, -1, or 0
current_position = 0
#record our trades
position_hist = []

#define the triggers
entry = 1.5
exit = 0.0

for index, row in df_trade.iterrows():

    z_score = row['Z_score']
    
    #check to see if state is safe or risky
    if row['State'] == 1:
        # get out if the position is risky
        current_position = 0
        position_hist.append(0)
        continue
    if row['State'] == 0:
    # if we own a position, and z_score goes above 0, then we sell for profit
        if current_position == 1:
            if z_score >= exit:
                current_position = 0
    # if we are currently short
        elif current_position == -1:
            if z_score <= -exit:
                current_position = 0
    # if we have no holdings
        else:
            # long entry, spread is too low
            if z_score < -entry:
                current_position = 1
            # Short entry, spread is too high
            elif z_score > entry:
                current_position = -1

    #record our position at the end of the day
    position_hist.append(current_position)


# save history
df_trade['Position'] = position_hist        
```

## Calculate Profit

```python
#determine spread movement
df_trade['Spread_Change'] = df_trade['Spread'].diff()

# if we held a position yesterday that moved in our favor today = more money
df_trade['Strategy_Return'] = df_trade['Position'].shift(1) * df_trade['Spread_Change']

# initial cash (easily changeable)
initial_cash = 10000

# sum all log-returns
df_trade['Culmulative_Return'] = df_trade['Strategy_Return'].cumsum()

# remove log
df_trade['Equity'] = initial_cash*np.exp(df_trade['Culmulative_Return'])

# add initial sum of cash
df_trade['Equity'] = df_trade['Equity'].fillna(initial_cash)

# visualize
plt.figure(figsize=(12, 6))
plt.plot(df_trade['Equity'], label='HMM Strategy', color='green', linewidth=2)
plt.title(f"Backtest Performance: BLK vs SYF (Start: ${initial_cash})")
plt.ylabel("Account Balance ($)")
plt.legend()
plt.grid(True, alpha=0.3)

final_balance = df_trade['Equity'].iloc[-1]
total_return_pct = (final_balance / initial_cash - 1) * 100


# Sharpe Ratio assuming we trade 252 days in a year
daily_std = df_trade['Strategy_Return'].std()
if daily_std == 0:
    sharpe=0
else:
    sharpe = (df_trade['Strategy_Return'].mean() / daily_std) * np.sqrt(252)

# measure the percent increased and decreased each day
rolling_max = df_trade['Equity'].cummax()
daily_drawdown = df_trade['Equity'] / rolling_max - 1
max_drawdown_pct = daily_drawdown.min() * 100

print("-" * 30)
print(f"Final Balance: ${final_balance:.2f}")
print(f"Total Return:  {total_return_pct:.2f}%")
print(f"Sharpe Ratio:  {sharpe:.2f}")
print(f"Max Drawdown:  {max_drawdown_pct:.2f}%")
print("-" * 30)
```

## Adjust Entry and Exit

```python
import seaborn as sns
import itertools
```

```python
Entry = np.linspace(0.5,2.5,21)
Exit = np.linspace(-0.5,1,16)
```

```python
# Make unique Entry and Exit pairs
unique_pairs = list(itertools.product(Entry,Exit))
```

```python
# find pairs where Entry is higher than Exit
valid_pairs = []
for pair in unique_pairs:
    if pair[0] > pair[1]:
        valid_pairs.append(pair)
```

```python
results = []
for pair in valid_pairs:
    # iterate through every day to make a decision
    #position can be 1, -1, or 0
    current_position = 0
    #record our trades
    position_hist = []
    
    #define the triggers
    entry = pair[0]
    exit = pair[1]
    
    for index, row in df_trade.iterrows():
    
        z_score = row['Z_score']
        
        #check to see if state is safe or risky
        if row['State'] == 1:
            # get out if the position is risky
            current_position = 0
            position_hist.append(0)
            continue
        if row['State'] == 0:
        # if we own a position, and z_score goes above 0, then we sell for profit
            if current_position == 1:
                if z_score >= exit:
                    current_position = 0
        # if we are currently short
            elif current_position == -1:
                if z_score <= -exit:
                    current_position = 0
        # if we have no holdings
            else:
                # long entry, spread is too low
                if z_score < -entry:
                    current_position = 1
                # Short entry, spread is too high
                elif z_score > entry:
                    current_position = -1
    
        #record our position at the end of the day
        position_hist.append(current_position)
    
    # save history
    df_trade['Position'] = position_hist 
    # if we held a position yesterday that moved in our favor today = more money
    df_trade['Strategy_Return'] = df_trade['Position'].shift(1) * df_trade['Spread_Change']

    # initial cash (easily changeable)
    initial_cash = 10000

    # sum all log-returns
    df_trade['Culmulative_Return'] = df_trade['Strategy_Return'].cumsum()

    # remove log
    df_trade['Equity'] = initial_cash*np.exp(df_trade['Culmulative_Return'])

    # add initial sum of cash
    df_trade['Equity'] = df_trade['Equity'].fillna(initial_cash)   

    final_balance = df_trade['Equity'].iloc[-1]

    # Sharpe Ratio assuming we trade 252 days in a year
    daily_std = df_trade['Strategy_Return'].std()
    if daily_std == 0:
       sharpe=0
    else:
        sharpe = (df_trade['Strategy_Return'].mean() / daily_std) * np.sqrt(252)

    # measure the percent increased and decreased each day
    rolling_max = df_trade['Equity'].cummax()
    daily_drawdown = df_trade['Equity'] / rolling_max - 1
    max_drawdown_pct = daily_drawdown.min() * 100

    # measure annualized return for Calmar ratio
    total_days = len(df_trade)
    years = total_days / 252

    # calculate compound annual growth rate
    total_return_ratio = final_balance/initial_cash
    cagr = (total_return_ratio ** (1/years)) - 1
    
    if max_drawdown_pct == 0:
            calmar = 0
    elif cagr < 0:
        calmar = -1 # Penalize losing strategies
    else:
            # Calmar = Total Return / |Max Drawdown|
            # (Using absolute value of max drawdown so the ratio is positive)
        calmar = cagr / abs(max_drawdown_pct/100)
            
    data = {
        'Entry':entry,
        'Exit':exit,
        'Sharpe':sharpe,
        'Final Balance':final_balance,
        'Max Drawdown %':max_drawdown_pct,
        'Calmar':calmar
    }

    results.append(data)
```

```python
df_results = pd.DataFrame(results)
df_results.head()
```
