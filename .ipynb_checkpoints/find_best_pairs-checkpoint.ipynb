## Find the best cointegration pairs for future trading

```python
import pandas as pd
import numpy as np
from hmmlearn.hmm import GaussianHMM
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# read in necessary csv
df_results = pd.read_csv("cointegration_results.csv")
df_prices = pd.read_csv("final_price_clean.csv", index_col=0, parse_dates=True)
```

```python
portfolio_candidates = []
all_equity_curves = pd.DataFrame()

for pair in df_results.head(20).itertuples():
    #calculate spread
    data_pair = df_prices[[pair[2],pair[3]]].dropna()
    x = np.log(data_pair.iloc[:,0])
    y = np.log(data_pair.iloc[:,1])
    y_c = sm.add_constant(y)
    model = sm.OLS(x,y_c).fit()
    spread = model.resid.values.reshape(-1,1)

    #Making the model
    hmm_model = GaussianHMM(n_components=2, n_iter=10000, 
                            min_covar=1e-3,random_state=42)
    try:
        hmm_model.fit(spread)
        hidden_states = hmm_model.predict(spread)
    except:
        print(f"  > Warning: HMM failed to converge for {pair[2]}-{pair[3]}. Skipping.")
        continue

    #determine which state is safe
    state_data = []
    for i in range(0,2):
        state_spreads = spread[hidden_states == i]
        safe_unsafe = {
            'State':i,
            'Variance':state_spreads.var()
        }
        state_data.append(safe_unsafe)
    risky_state_info = max(state_data, key=lambda x: x['Variance'])
    risky_state = risky_state_info['State']
    safe_state_info = min(state_data, key=lambda x: x['Variance'])
    safe_state = safe_state_info['State']

    # calculate z-score & create trading dataframe for future algorithm
    z_score = (spread - spread.mean()) / spread.std()
    df_trade = pd.DataFrame(index = data_pair.index)
    df_trade['Spread'] = spread
    df_trade['State'] = hidden_states
    df_trade['Z_score'] = z_score

    current_position = 0
    #record our trades
    position_hist = []
    #use previously found entry/exit
    entry = 0.9
    exit = -0.5
    for index, row in df_trade.iterrows():
        z_score = row['Z_score']
        #check to see if state is safe or risky
        if row['State'] == risky_state:
            # get out if the position is risky
            current_position = 0
            position_hist.append(0)
            continue
        if row['State'] == safe_state:
        # if we own a position, and z_score goes above 0, then we sell for profit
            if current_position == 1:
                if z_score >= exit:
                    current_position = 0
        # if we are currently short
            elif current_position == -1:
                if z_score <= -exit:
                    current_position = 0
        # if we have no holdings
            else:
                # long entry, spread is too low
                if z_score < -entry:
                    current_position = 1
                # Short entry, spread is too high
                elif z_score > entry:
                    current_position = -1
    
        #record our position at the end of the day
        position_hist.append(current_position)


    # save history
    df_trade['Position'] = position_hist

    #determine spread movement
    df_trade['Spread_Change'] = df_trade['Spread'].diff()
    # if we held a position yesterday that moved in our favor today = more money
    df_trade['Strategy_Return'] = df_trade['Position'].shift(1) * df_trade['Spread_Change']
    initial_cash = 10000
    # sum all log-returns
    df_trade['Culmulative_Return'] = df_trade['Strategy_Return'].cumsum()
    # remove log
    df_trade['Equity'] = initial_cash*np.exp(df_trade['Culmulative_Return'])
    # add initial sum of cash
    df_trade['Equity'] = df_trade['Equity'].fillna(initial_cash)   
    final_balance = df_trade['Equity'].iloc[-1]

    # Sharpe Ratio assuming we trade 252 days in a year
    daily_std = df_trade['Strategy_Return'].std()
    if daily_std == 0:
       sharpe=0
    else:
        sharpe = (df_trade['Strategy_Return'].mean() / daily_std) * np.sqrt(252)
    
    # measure the percent increased and decreased each day
    rolling_max = df_trade['Equity'].cummax()
    daily_drawdown = df_trade['Equity'] / rolling_max - 1
    max_drawdown_pct = daily_drawdown.min() * 100
    
    # measure annualized return for Calmar ratio
    total_days = len(df_trade)
    years = total_days / 252
    
    # calculate compound annual growth rate
    total_return_ratio = final_balance/initial_cash
    cagr = (total_return_ratio ** (1/years)) - 1

    if max_drawdown_pct == 0:
        calmar = 0
    elif cagr < 0:
        calmar = -1 
    else:
        calmar = cagr / abs(max_drawdown_pct/100)

    if sharpe > 1.0 and max_drawdown_pct > -25:
        candidate_info = {
            'Stock1':pair[2],
            'Stock2':pair[3],
            'Sharpe':sharpe,
            'Drawdown':max_drawdown_pct,
            'Final Balance':final_balance
        }
        portfolio_candidates.append(candidate_info)

        pair_name = f"{pair[2]}-{pair[3]}"
        all_equity_curves[pair_name] = df_trade['Equity']

# visualizing everything
if not all_equity_curves.empty:
    # Fill missing data with starting cash
    all_equity_curves = all_equity_curves.fillna(initial_cash)
    
    # Calculate Total Portfolio Value
    # Sum across all columns (axis=1) to get the "Fund Value" for each day
    all_equity_curves['Total_Fund'] = all_equity_curves.sum(axis=1)
    
    plt.figure(figsize=(12, 6))
    plt.plot(all_equity_curves['Total_Fund'], label='Total Hedge Fund Value', color='green', linewidth=2.5)
    
    for column in all_equity_curves.drop(columns=['Total_Fund']).columns:
        plt.plot(all_equity_curves[column], alpha=0.2, linewidth=1) # Faint lines
        
    plt.title(f"Hedge Fund Performance (8 Uncorrelated Strategies)")
    plt.ylabel("Total Portfolio Value ($)")
    plt.xlabel("Date")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    # Final metrix
    initial_investment = 10000 * len(portfolio_candidates) # e.g., $80,000
    final_fund_value = all_equity_curves['Total_Fund'].iloc[-1]
    total_fund_return = ((final_fund_value / initial_investment) - 1) * 100
    
    print(f"Initial Investment: ${initial_investment:,.2f}")
    print(f"Final Fund Value:   ${final_fund_value:,.2f}")
    print(f"Total Fund Return:  {total_fund_return:.2f}%")
```

```python
df_portfolio = pd.DataFrame(portfolio_candidates)
    
# sort by highest Sharpe
df_portfolio = df_portfolio.sort_values(by='Sharpe', ascending=False)

# Print the leaderboard
print(df_portfolio[['Stock1', 'Stock2', 'Sharpe', 'Drawdown', 'Final Balance']].to_string(index=False))

# Calculate metrics
avg_sharpe = df_portfolio['Sharpe'].mean()
avg_drawdown = df_portfolio['Drawdown'].mean()
total_expected_return = df_portfolio['Final Balance'].sum() - (10000 * len(df_portfolio))

print("-" * 40)
print(f"Portfolio Average Sharpe:   {avg_sharpe:.2f}")
print(f"Portfolio Average Drawdown: {avg_drawdown:.2f}%")
print(f"Total Theoretical Profit:   ${total_expected_return:,.2f}")
print("-" * 40)
```
